## USA

```python
feb_usa= (0.0*29)/29
march_usa= (2.0*16)/31
april_usa= (2.0*30)/30
may_usa= (2.0*31)/31
jun_usa= (2.0*30)/30
jul_usa= (2.0*19)
jul_usa2= (1.0*12)
jultotal_usa= (jul_usa+jul_usa2)/31
aug_usa= (1.0*31)/31
sep_usa= (1.0*11)
sep_usa2= (1.0*19)
septotal_usa= (sep_usa+sep_usa2)/30

months_usa = [feb_usa, march_usa, april_usa, may_usa, jun_usa, jultotal_usa, aug_usa, septotal_usa]
month_usa_value = pd.DataFrame(months_usa)

months_us= {'CountryName': ['USA','USA','USA','USA','USA','USA','USA','USA'],
        'Year_Month': ['2020-02', '2020-03', '2020-04', '2020-05', '2020-06', '2020-07', '2020-08', '2020-09']}
usa = pd.DataFrame(months_us, columns=['CountryName', 'Year_Month'])
usa['PolicyValue'] = month_usa_value
```

### Concat countries, turn it into a new dataframe, and reset index

```python
#Put all dataframes together (concat) and turn them into a new dataframe
prep = pd.concat([brazil, italy, singapore, safa, usa])
prep_df = pd.DataFrame(prep)

#Resent the Index
prep_df.index = range(len(prep_df))
prep_df

```
