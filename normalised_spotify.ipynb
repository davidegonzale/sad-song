### BRAZIL

```python
df = pd.DataFrame(data)
grouped_bra =df[df['country']== '-za-'].groupby(['date'])['date','country','Streams','danceability','energy','valence','tempo'].mean()
grouped_bra['HappyIndex'] = (grouped_bra.energy*grouped_bra.valence*grouped_bra.Streams)/grouped_bra.Streams
grouped_bra

#grouped2 = grouped[(grouped ['date'] >'2019-12-15')]
```

```python
brazil_index= grouped_bra.iloc[10:18, 5:6]
brazil_index= brazil_index.reset_index()
brazil_index
```

```python
def year_month(string):
    return string[0:7]

brazil_index['Year']= brazil_index['date'].apply(year_month)
brazil_index   
```

```python
brazil_index2 = brazil_index[['Year', 'HappyIndex']]
brazil_index2
```

```python
months_brazil= {'CountryName': ['Brazil','Brazil','Brazil','Brazil','Brazil','Brazil','Brazil','Brazil']}
brazil = pd.DataFrame(months_brazil, columns=['CountryName'])
brazil['Year_Month']= brazil_index2['Year']
brazil['Happy Index']= brazil_index2['HappyIndex']
#brazil
```
